```python
import pandas as pd

data_path = "/Users/philippjohn/Developer/youtube-analytics-data/"
car = "Porsche_911_GT3â€‹"

# New df which filters to only include comments that contain meantion of a feature
def get_df_with_feature(car, df, feature):
    df_feature = pd.DataFrame()
    df_feature = pd.concat([df_feature, df[df["content_clean"].str.lower().str.contains(feature)]], axis=0)
    df_feature["feature"] = feature
    return df_feature

# Read csv df
df = pd.read_csv(data_path + car + "/sentiment_1.csv", header=[0], lineterminator='\n')
df = df.drop(['Unnamed: 0'], axis=1, errors='ignore')

# New df which filters to only include comments that contain meantion of a feature
features = ["rim", "steering wheel", "engine", "color", "colour", 
            "carbon", "light", "design", "sound", "interior", "exterior", 
            "mirror", "body", "brake", "chassis", "suspension", "gearbox", 
            "navigation", "infotainment"]
df_features = pd.DataFrame()
for feature in features:
    df_features = pd.concat([df_features, get_df_with_feature(car, df, feature)], axis=0)
df_features.head()
```

```python
df_feature_stats = []
for feature in features:
    df_feature_stats.append([feature,
                         len(df_features[df_features["feature"] == feature]),
                         df_features[df_features["feature"] == feature]["sentiment"].mean()])

df_feature_stats = pd.DataFrame(df_feature_stats, columns=['feature', 'comment_count', 'sentiment_mean'])
df_feature_stats = df_feature_stats[df_feature_stats['comment_count'] >= 5]  
df_feature_stats = df_feature_stats.sort_values(by=["sentiment_mean"], ascending=False)
df_feature_stats = df_feature_stats.reset_index(drop=True)
df_feature_stats
```
